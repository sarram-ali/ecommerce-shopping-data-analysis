```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('shopping_trends.csv')
# add colors for future use
colors = ["#2E86C1","#48C9B0","#AF7AC5","#F4D03F","#EB984E","#EC7063","#5D6D7E","#82E0AA"]
```

## ðŸ’° 1. What is Our Total Revenue?

    Total revenue (sum of Purchase Amount)

    Average Purchase Amount

    Minimum Purchase Amount

    Maximum Purchase Amount

    Create a Simple Bar Chart Showing Min, Avg, Max

```python
total_revenue = df['Purchase Amount (USD)'].sum()
average_purchase = df['Purchase Amount (USD)'].mean()
min_purchase = df['Purchase Amount (USD)'].min()
max_purchase = df['Purchase Amount (USD)'].max()

print("Total Revenue:", total_revenue)
print("Average Purchase:", average_purchase)
print("Minimum Purchase:", min_purchase)
print("Maximum Purchase:", max_purchase)
```

```python
details = df['Purchase Amount (USD)'].describe()
details
```

```python
details.drop(['count','std','25%','50%','75%'],inplace=True)
```

```python
fig = go.Figure(go.Bar(x=details.index,y=details.values))
fig.update_layout(title = 'Purchase Amount Details',xaxis_title = 'Maximum, Minimum, Average Revenue',yaxis_title = 'Price')
fig.add_annotation(x='max',y=max_purchase,text='Highest',showarrow=True,arrowhead=5,arrowwidth=2,ax=20,ay=-50)
fig.show()
```

## 2. Which Product Category Generates the Highest Total Revenue?

    Total revenue per Category

    Top category

    Bar chart of all categories (sorted descending)

```python
cate = df.groupby('Category')['Purchase Amount (USD)'].sum()
cate_max = df.groupby('Category')['Purchase Amount (USD)'].sum().max()
cate_max
```

```python

cate = cate.sort_values(ascending=False)
```

```python
fig = go.Figure(go.Bar(x=cate.index,y=cate.values))
fig.update_layout(title = 'Purchase Amount Details',xaxis_title = 'Category',yaxis_title = 'Price')
fig.add_annotation(x='Clothing',y=cate_max,text='Highest',showarrow=True,arrowhead=5,arrowwidth=2,ax=20,ay=-50)
fig.show()
```

## 3. ðŸ—º Which state generates the most revenue?


    Calculate total revenue

    Show top 5 states

    Create visualization

```python
all_state_revenue = df.groupby('Location')['Purchase Amount (USD)'].sum()
```

```python
top_5 = all_state_revenue.sort_values(ascending=False).head(5)
top_5
```

```python
fig = go.Figure(go.Pie(labels=top_5.index,values=top_5.values,hole=0.4,text=top_5.index,marker=dict(colors = colors),
    textfont=dict(color='black')))
fig.update_layout(title = 'Top 5 States Generates Most Revenue')
fig.show()
```

## 5.ðŸŽ¯ Does Age affect spending?

    Create age groups (18â€“25, 26â€“35, 36â€“45, 46â€“55, 56+)

    Total revenue per age group

    Average spending per age group

    Visualization

```python
bins = [18,25,35,45,55,100]
labels = ['18-25','26-35','36-45','46-55','56+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
age_revenue = df.groupby('Age Group')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
age_revenue
```

```python
age_avg = df.groupby('Age Group')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
age_avg
```

```python
fig = px.bar(age_revenue,title='Total Revenue by Age Group')
fig.update_layout(xaxis_title='Age Group',yaxis_title='Purchase Amount (USD)')
fig.show()
```

## 6.ðŸ› Which Product Category is most popular among each Gender?

    Group by Gender + Category

    Calculate total revenue

    Find top category for each gender

    Visualize

```python
cate_rev = df.groupby(['Gender','Category'])['Purchase Amount (USD)'].sum().reset_index()
cate_rev
```

```python
top_cate = cate_rev.sort_values(['Gender','Purchase Amount (USD)'], ascending=[True, False])
top_cate = top_cate.groupby('Gender').head(1)
print(top_cate)
```

```python
fig = px.bar(
    top_cate,
    x='Gender',
    y='Purchase Amount (USD)',
    color='Category',
    title='Top Revenue Category for Each Gender'
)

fig.show()
```
